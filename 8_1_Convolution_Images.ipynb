```python
from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu')
```

합성곱 계산은 (4,4) 크기의 입력에 (3,3) 크기의 커널을 적용해서 (2,2) 크기의 특성 맵을 만듬. 커널을 그대로 두고 출력의 크기를 입력과 동일하게 (4,4)로 만들려면????? 입력이 마치 (6,6)인거처럼 구라까기!!!!! ***padding***

Screenshot 2024-07-12 at 11.37.33 PM.png

**Same padding** : 입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩하는것.

- 입력과 특성 맵의 크기를 동일하게 만든다.

**Valid padding** : 패딩 없이 순수한 입력 배열에서만 합성곲을 하여 특성 맵을 만드는 경우.

- 특성 맵의 크기가 줄어든다.

패딩을 사용하는 이유.

픽셀마다 합성곱 연산에 기여하는 바가 달라 주변 픽셀들의 중요한 정보가 있다면 잘 감지하기 어렵다.

***Stride*** : 커널 이동의 크기. 보통 1. 잘 바꾸지 않음.

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same', strides=1)
```

***Pooling*** : 합성곱 층에서 만든 특성 맵의 크기를 줄이는 역할. 파라미터를 줄이기 때문에 가대적합을 방지. 계산이 줄어들어 리소스 절약. (2,2,3) >> (1,1,3). 풀링도 합성곱처럼 입력 위를 지나가면서 도장을 찍음.

**Max Pooling** : 도장을 찍은 영역에서 가장 큰 값을 고름.

**Average Pooling** : 평균값을 고름.

풀링은 합성곱 층과 뚜렷이 구분되기 때문에 풀링 층이라고 부름.

스트라이드는 풀링의 크기에 따라. 겹치지 않게하기위해서 풀링의 크기가 (2,2)이면 스트라이드 2. (3,3)이면 3.

Screenshot 2024-07-13 at 12.16.23 AM.png

```python
keras.layers.MaxPooling2D(2) #2: 풀링의 크기
```

```python
keras.layers.MaxPooling2D(2, strides=2,  padding='valid') #풀링은 패딩을 하지 않는다. 이 매개변수 바꾸는 경우 거의 없음.
```
