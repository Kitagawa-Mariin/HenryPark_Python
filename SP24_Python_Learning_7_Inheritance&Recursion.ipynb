# 복습

class란?

class는 클론 생성 공장이라 생각하면 편하다. 각 클론마다 고유의 특징들이 있으며, 공장에서 부여해준 기능들도 탑재되어 있다.

이 고유 특성들을 self variable 혹은, object feature라고 부른다. 오브젝트란, 공장에서 만들어진 클론을 지칭한다.

오브젝트의 초기 값을 설정해주는 것이 "--init--" 함수이다. 이 함수에 내가 생성할 오브젝트의 특성들을 넣으면 그 특성을 가지고있는 클론이 많들어지는 것이다.

마지막으로 class에는 class method라고 함수들이 있는데, 이 함수들은 내 클론이 가지고있는 기능이라 생각하면 편하다.

펑션과 다른 점은 메소드는 클래스 오브젝트들만 사용할 수 있는 함수이다.



```python
class apt:
  def __init__(self, name, comp, floor, room): #초기값설정하는 것 무조건 class안에 __init__ 있어야됨
    self.name = name
    self.comp = comp
    self.floor = floor
    self.room = room

  def get_name(self):
    return self.name

  def get_comp(self):
    return self.comp

  def get_room(self):
    return self.room

  def change_room(self, room):
    self.room = room

  def print_feature(self):
    print(f"I am living at an apartment named {self.name} built by {self.comp}. My detailed adress is floor {self.floor}, {self.room}APT")
```

```python
hangoos_home = apt("샴바나래미안", "삼성", 99, 9909)

print(hangoos_home.get_name())
print(hangoos_home.get_comp())
print(hangoos_home.get_room())

hangoos_home.change_room(9902)
print(hangoos_home.get_room())
hangoos_home.print_feature()
```

# Inheritance 상속

상속이란, 부모 클래스로부터 모든 특징들과 메소드들을 다 물려받고 + 자신만의 feature와 method가 있는 그런 자식 클래스를 말합니다.

상속을 왜 쓰나요?

상속은 소규모 프로젝트에선 잘 쓰이지 않지만, 대규모 프로젝트에선 빈번하게 쓰입니다.

이미 존재하는 클래스의 특징들을 모두 가지고 있음과, 거기에 추가적으로 자신만의 기능들을 넣을 수 있다는 메리트가 대규모 프로젝트에선, 다양한 사람들이 각각 만든 클래스들을 가져다 쓸 수 있는 자유로움을 줍니다.

예를 들어 저와 러닝팀 모두가 함께 프로젝트를 진행하는데, 제가 school 클래스를 만들었다고 생각합시다.

여러분들은 제가 만든 학교 클래스에 이어, 그 학교에 다니는 학생들을 구현하기 위한 클래스를 만들고 싶습니다. 이 때 학생 클래스는 학교 클래스의 특징들까지 가지고 싶을 경우엔 제 학교 클래스를 상속하면 됩니다.

```python
class house(apt):
  def __init__(self, house_name, comp, floor, room, family_size, my_name):
    super().__init__(house_name, comp, floor, room) # 부모 클래스의 이니셜라이징 함수를 상속 + 모든 메소드들 또한 상속
    self.family_size = family_size # 자식 클래스만의 새로운 변수들
    self.my_name = my_name

  def get_my_name(self):
    return self.my_name


good_house = house("주택", "삼성", 99, 9909, 5, "강한구")

print(good_house.get_my_name())

print(good_house.get_name()) # 슈퍼로 상속받은 함수 사용 가능

good_house.print_feature()
```

# Recursion (재귀함수)

Recursion은 한국어로는 재귀함수라 불린다.

재귀함수는 함수 안에서 같은 함수를 또 부르는 것을 뜻한다.

재귀 함수가 자기 스스로를 반복해서 부르는 만큼, 언제 멈출지 "base case"를 정해줘야 합니다.

아래의 팩토리얼 재귀함수를 예시로 보겠습니다.

```python
def factorial(n):
  if n == 1: # base case
    return 1 # when n == 1 just return 1 to stop
  return n * factorial(n - 1) # when n != 1, do n * factorial n -1

print(factorial(5))
```

팩토리얼 재귀함수 같은 경우엔 처음에 n이 들어가고,

$$ n * factorial(n - 1) $$

을 하면서 반복해서 곱해줍니다. 이럴 경우엔 새로운 팩토리얼 함수에 n-1 값이 들어가게 되고, 그러면 n-1 * fac(n - 2)가 계산되게 됩니다,

다시, fac(n-2)는 또 n-2 * fac(n - 3)을 계산하고, 이를 n == 1이 될 때까지 반복합니다.

그러면 결국엔,

$$n * (n-1) * (n -2) ... * 1$$

이 계산 됩니다.

재귀함수는 왜 쓰나요?

연속적인 함수의 부름이 필요한 알고리즘에서 유용합니다. 특히나 어떠한 큰 문제를 작은 문제들로 나눠서 풀 때 유용합니다. (Divide-and-conquer)

## 예제 1

피보나치 함수를 재귀함수를 사용하여 구현하시오.

피보나치 함수란? n = n-1 + n-2의 값을 가지는 숫자를 말합니다.

image.png

```python
def fibonacci(n):
  if n == 1 or n == 2:
    return 1
  return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```

# 예제 2

pow(2,4)라는 함수를 만들어봅시다

파이썬에서 간단히 2 ** 4 라는 방법을 사용할 수 있지만

이번에 배운 recursion와 함수를 사용하여 똑같은 기능을 구현하도록 만들어 봅시다.

hint) a 와 b가 각각 0일때 어떻게 다루어야하는지 생각해봅시다.

```python
def pow(a,b):
  if b == 0:
    return 1
  if a == 0:
    return 0
  return a*pow(a,b-1)
assert pow(2,4) == 16
assert pow(0,0) == 1
assert pow(0,3) == 0
assert pow(1,0) == 1
assert pow(3,4) == 81
```
