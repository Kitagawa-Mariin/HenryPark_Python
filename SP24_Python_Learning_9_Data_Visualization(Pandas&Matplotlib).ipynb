# Data Visualization

## Introduction

**Pandas**는 다양한 데이터 분석 업무에서 사용되는 패키지입니다.

넘파이 때와 같이 판다스는 임포트 해야합니다.

```python
# We use alias 'pd' when using pandas
import pandas as pd
```

**Matplotlib**은 데이터 시각화를 도와주는 패키지입니다.

To use matplotlib, we need to write this code snippet!

```python
# We use alias 'plt' when using matplotlib
from matplotlib import pyplot as plt
```

다양한 데이터 파일의 형식이 있으나, 대다수의 데이터가 cvs파일 혹은 JSON 파일로 제공이 될 겁니다. 이럴 때는 pandas의 **read_csv**라는 함수를 사용하여 데이터를 불러오면 됩니다.

```python
# we are reading aapl.csv file
# apple = pd.read_csv("https://gist.github.com/netj/8836201#file-iris-csv")
# apple

url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
iris = pd.read_csv(url)

iris.head()
```

info 함수를 이용하여 데이터에 널 값이 뭐가 있는지, 어떤 클래스가 있는지 볼 수 있습니다.

```python
iris.info()
```

**iloc**함수를 이용하여 특정 행의 특정 열 값을 불러올 수 있습니다.

## Using pandas...

```python
# data in 5th row, 1st column
iris.iloc[4, 0]
```

```python
# data in 2nd to 4th row, 4th column
iris.iloc[1:4, 3]
```

```python
# data in 2nd to 4th row, 1st to 4th column
iris.iloc[1:4, 0:4]
```

```python
# data in last to last + 4th row, last column
iris.iloc[-4:, -1]
```

In addition to connecting using coordinates, you can also connect using **Column type name**!

```python
# getting all data under 'variety' column
iris['variety']
```

**loc**은 특정 class 밑의 모든 행 혹은 열을 보여줍니다.

```python
# data in first row, under column 'variety'
iris.loc[1, 'variety']
```

We can use **mean()** function to calculate the mean of data in the given list.

```python
# Mean of data from 2nd to 5th row, under third column.
iris.iloc[1:4, 3].mean()
iris['sepal.length'].mean()
```

To filter output data, we can...

```python
# Inside the bracket that is used for indexing, we can put condition.
iris[iris['variety'] == "Setosa"]
```

```python
iris.head()
```

Can you tell the difference between the two code snippets below?

```python
iris[(iris['sepal.length'] > 5.0) & (iris['sepal.width'] < 3.2)]["variety"]
```

## Data visualization

By using the matplotlib library, we can create graphs!

```python
# Plotting values under column 'Close' as lines
plt.plot(iris['sepal.length'])
```

If you set the graph to match a given date (with the date on the x-axis), a visual error will occur.

```python
plt.plot(iris['sepal.length'], iris['sepal.width'])
```

Let's look at the functions of matplotlib.pyplot while looking at the detailed process of creating a graph.

```python
# Setting the size of the graph
plt.figure(figsize = (20, 10))

# Creating a graph by setting the x and y axes
plt.plot(iris['sepal.length'], iris['petal.length'])

# Creating a title to be placed at the top of the graph
plt.title('sepal length vs petal length')

# Specify a name for the x-axis of the graph
plt.xlabel("length")

# Specify a name for the y-axis of the graph
plt.ylabel("something")
```
